```python
import zipfile
import tarfile
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def plot_sample(img1, img2, img3, img4):
    image1 = plt.imread(img1)
    image2 = plt.imread(img2)
    image3 = plt.imread(img3)
    image4 = plt.imread(img4)
    
    plt.subplot(2, 2, 1)
    plt.imshow(image1)
    plt.title(img1.split('/')[-1])
    
    
    plt.subplot(2, 2, 2)
    plt.imshow(image2)
    plt.title(img2.split('/')[-1])
    
    plt.subplot(2, 2, 3)
    plt.imshow(image3)
    plt.title(img3.split('/')[-1])

    plt.subplot(2, 2, 4)
    plt.imshow(image4)
    plt.title(img4.split('/')[-1])
    
    plt.show()
```
